```python
from selenium import webdriver
import lxml.html as lh
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path='G:/chromedriver.exe')
URL = 'https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=24&Tx_State=UP&Tx_District=1&Tx_Market=0&DateFrom=01-Jan-2020&DateTo=31-Dec-2020&Fr_Date=01-Jan-2020&To_Date=31-Dec-2020&Tx_Trend=0&Tx_CommodityHead=Potato&Tx_StateHead=Uttar+Pradesh&Tx_DistrictHead=Agra&Tx_MarketHead=--Select--'
driver.get(URL)
driver.implicitly_wait(5)

```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
dfs=[]
for i in range(round(1827/50)):
    soup = BeautifulSoup(driver.page_source, 'lxml')

    tables = soup.find('table',id="cphBody_GridPriceData")

    # Step 3: Read tables with Pandas read_html()
    dfs .append( pd.read_html(str(tables)))
    print(dfs[i])
    
    if i==0 :
        xpath_first_view =  '//*[@id="cphBody_GridPriceData"]/tbody/tr[52]/td/table/tbody/tr/td[1]/input'
    
        time.sleep(1)
        next_btn = driver.find_element_by_xpath(xpath_first_view).click()
        time.sleep(10)
    elif i==round(1827/50)-1:
        driver.close()
    else :
        xpath_first_view =  '//*[@id="cphBody_GridPriceData"]/tbody/tr[52]/td/table/tbody/tr/td[3]/input'
   
        time.sleep(1) 
        next_btn = driver.find_element_by_xpath(xpath_first_view).click()
        time.sleep(10)
    i=i+1
```

### Loading the data from the list

```python
data=pd.DataFrame()
for i in range(len(dfs)):
    data=pd.concat([data,dfs[i][0]],axis=0)
data
```

```python
data=data.dropna()
data=data.rename(columns={'Sl no.': 'S_Number'})
data=data.reset_index(drop=True)
data.head()
```
